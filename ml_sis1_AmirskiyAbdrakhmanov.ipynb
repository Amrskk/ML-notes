```python
# Mode imputation
df_stat = df.copy()

df_stat['Distance_from_Home'] = df_stat['Distance_from_Home'].fillna(df_stat['Distance_from_Home'].mode()[0])
df_stat['Teacher_Quality'] = df_stat['Teacher_Quality'].fillna(df_stat['Teacher_Quality'].mode()[0])
df_stat['Parental_Education_Level'] = df_stat['Parental_Education_Level'].fillna(df_stat['Parental_Education_Level'].mode()[0])

display(df_stat.isna().sum())
```

```python
# I  don't have missing numerics in the dataset !!
df_stat1 = df.copy()

artif_cols = ['Hours_Studied', 'Exam_Score']
n_rows = len(df_stat1)
nan_propor = 0.1
nans = int(n_rows * nan_propor)

randoms = np.random.choice(df_stat1.index, size=nans, replace=False)
df_stat1.loc[randoms, artif_cols] = np.nan

display(df_stat1.isna().sum())
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder

df_encod = df.copy()


ordinal_feats = ['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home']
ordinal_cats = [
    ['Low', 'Medium', 'High'],
    ['High School', 'College', 'Postgraduate'],
    ['Near', 'Moderate', 'Far']
]
nominal_feats = [c for c in cat_features if c not in ordinal_feats]

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
oe = OrdinalEncoder(
    categories=ordinal_cats,
    handle_unknown='use_encoded_value',
    unknown_value=np.nan
)
df_encod[ordinal_feats] = oe.fit_transform(df_encod[ordinal_feats])
nominal_encoded = ohe.fit_transform(df_encod[nominal_feats])
nominal_column_names = ohe.get_feature_names_out(nominal_feats)
df_numeric = df_encod.drop(columns=nominal_feats)
df_final_input = pd.concat([df_numeric.reset_index(drop=True), pd.DataFrame(nominal_encoded, columns=nominal_column_names)], axis=1)

imputer = IterativeImputer(random_state=42)
df_ii_imputed = imputer.fit_transform(df_final_input)
df_predimp = pd.DataFrame(df_ii_imputed, columns=df_final_input.columns)

print(df_predimp.isna().sum())
```

##### Constant Value Imputation
This involves filling missing slots with a fixed value (e.g., 0, "Missing", or "Unknown").

Advantages:

1)Simplicity. Extremely fast and easy to implement.

2)Preservation. By using a label like "Unknown," you preserve the information that the data was missing, which might be a meaningful pattern in itself.

Disadvantages:

1)Bias: Can significantly distort the mean and median if the constant (like 0) isn't representative of the true data.

2)Noise: Introduces "fake" spikes in the data that can mislead models.


##### Mean/Median/Mode Imputation
This uses the central tendency measures of the data to fill the gaps.

Advantages:

1)Efficiency: Very fast and works well with small datasets where missingness is completely random.

2)Consistency: Mean imputation keeps the sample mean unchanged.

Disadvantages:

1)Underestimates Variance: It artificially shrinks the Standard Deviation because you are adding many values exactly at the center.


##### Predictive Imputation 

Advantages

1)Captures Relationships: Unlike simple mean/mode imputation, it models each feature as a function of others.

2)Maintains Statistical Variance: It preserves the distribution and relationships in your dataset better than "filling with zeros" or "filling with the average," which often "blunts" the data.

3)Flexibility: Since it uses a regression-style approach (by default BayesianRidge), it can handle a mix of different data types once they are properly encoded.

4)Reduced Bias: It generally leads to less biased estimates in downstream machine learning models compared to simpler methods.

Disadvantages

1)Computational Expense: It is much slower than simple imputation because it runs a loop of regressions across all columns multiple times. This can be a bottleneck for very large datasets.

2)Potential for Leakage: If not used carefully within a Scikit-learn Pipeline, you risk "data leakage" (information from your test set leaking into your training set). But here I didn't do anything complex so all good


###### Min-Max Normalization [0,1] - When to Use:

**Best suited for:**

 **Algorithms sensitive to feature (distance?)**:
- Distance-based algorithms: KNN, K-Means clustering, hierarchical clustering
- Tree-based models don't require scaling, but ensemble methods benefit when features are normalized
- These algorithms calculate distances between data points, so the actual scale matters

**Advantages:**
- Simple and intuitive
- Preserves the shape of original distribution
- Values are bounded to [0,1], easy to understand
- Suitable for some algorithms like neural networks with sigmoid activation

**Disadvantages:**
- Sensitive to outliers (outliers can compress other values toward 0)
- If new data exceeds original min/max, it will be scaled outside [0,1]
- Changes the variance differently for different features

---

###### Z-Score Standardization - When to Use:

**Best suited for:**

**Assuming normal distribution**: When your data approximately follows a normal distribution
   - Linear regression, logistic regression
   - Gaussian Naive Bayes
   - Principal Component Analysis (PCA)

**Statistical methods**: Any algorithm based on statistical assumptions
   - t-tests, ANOVA, and other hypothesis tests
   - Linear/Ridge/Lasso regression - these methods assume normally distributed residuals

 **Outlier-robust scaling**: When you have outliers that shouldn't be compressed
   - Z-score standardization is less affected by extreme values
   - Outliers will have large |z| values but won't compress other values

 **Machine Learning models**: Most sklearn models perform better with z-score standardization:
   - Support Vector Machines (SVM)
   - Linear models with regularization



**Advantages:**
- Less sensitive to outliers than min-max
- Mathematically convinient - mean=0, std=1
- Suitable for statistical tests and regularization
- Works well with unbounded ranges
- Better for algorithms based on statistical assumptions

**Disadvantages:**
- Less intuitive (hard to interpret what z=1.5 means)
- Not bounded - values can exceed [-3, 3]
- Assumes approximately normal distribution
- Extreme outliers will have very large z-scores


#### 2.2 B2. Feature Creation & Discretization

```python
df_engineered = df.copy()

df_engineered['Study_Efficiency'] = df_engineered['Exam_Score'] / df_engineered['Hours_Studied']

df_engineered['Study_Load'] = (df_engineered['Hours_Studied'] * 0.5 + 
                               df_engineered['Tutoring_Sessions'] * 2 + 
                               df_engineered['Attendance'] * 0.3)

df_engineered['Preparedness'] = (df_engineered['Previous_Scores'] * 0.6 + 
                                 df_engineered['Hours_Studied'] * 5)


print("\n1. Study_Efficiency (Exam_Score / Hours_Studied):")
print(f"   Min: {df_engineered['Study_Efficiency'].min():.2f}")
print(f"   Max: {df_engineered['Study_Efficiency'].max():.2f}")
print(f"   Mean: {df_engineered['Study_Efficiency'].mean():.4f}")
print(f"   Std: {df_engineered['Study_Efficiency'].std():.4f}")

print("\n2. Study_Load (weighted combination):")
print(f"   Min: {df_engineered['Study_Load'].min():.2f}")
print(f"   Max: {df_engineered['Study_Load'].max():.2f}")
print(f"   Mean: {df_engineered['Study_Load'].mean():.4f}")
print(f"   Std: {df_engineered['Study_Load'].std():.4f}")

print("\n3. Preparedness (weighted score):")
print(f"   Min: {df_engineered['Preparedness'].min():.2f}")
print(f"   Max: {df_engineered['Preparedness'].max():.2f}")
print(f"   Mean: {df_engineered['Preparedness'].mean():.4f}")
print(f"   Std: {df_engineered['Preparedness'].std():.4f}")

print("\nFirst 10 rows with engineered features:")
display(df_engineered[['Hours_Studied', 'Exam_Score', 'Study_Efficiency', 'Study_Load', 'Preparedness']].head(10))
```

##### Discretization - Equal-Width Binning

 Bin width = (max - min) / k


```python
n_bins = 5

df_engineered['Study_Efficiency_EqualWidth'] = pd.cut(
    df_engineered['Study_Efficiency'], 
    bins=n_bins, 
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)

df_engineered['Study_Load_EqualWidth'] = pd.cut(
    df_engineered['Study_Load'], 
    bins=n_bins,
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)

print("\nStudy_Efficiency - Equal-Width Bins:")
print(f"Number of bins: {n_bins}")
ew_efficiency_counts = df_engineered['Study_Efficiency_EqualWidth'].value_counts().sort_index()
print("\nBin Distribution:")
print(ew_efficiency_counts)

print("\n\nStudy_Load - Equal-Width Bins:")
ew_load_counts = df_engineered['Study_Load_EqualWidth'].value_counts().sort_index()
print("\nBin Distribution:")
print(ew_load_counts)

# Get bin edges
efficiency_bins = pd.cut(df_engineered['Study_Efficiency'], bins=n_bins, retbins=True)[1]
load_bins = pd.cut(df_engineered['Study_Load'], bins=n_bins, retbins=True)[1]

print("\n\nBin Boundaries (Equal-Width):")
print(f"\nStudy_Efficiency bins: {np.round(efficiency_bins, 3)}")
print(f"Study_Load bins: {np.round(load_bins, 3)}")
```

##### Discretization - Equal-Frequency Binning
Each bin contains roughly n/k samples


```python
df_engineered['Study_Efficiency_EqualFreq'] = pd.qcut(
    df_engineered['Study_Efficiency'], 
    q=n_bins, 
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
    duplicates='drop'
)

df_engineered['Study_Load_EqualFreq'] = pd.qcut(
    df_engineered['Study_Load'], 
    q=n_bins,
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
    duplicates='drop'
)


print("\nStudy_Efficiency - Equal-Frequency Bins:")
ef_efficiency_counts = df_engineered['Study_Efficiency_EqualFreq'].value_counts().sort_index()
print("\nBin Distribution (should be more balanced):")
print(ef_efficiency_counts)

print("\n\nStudy_Load - Equal-Frequency Bins:")
ef_load_counts = df_engineered['Study_Load_EqualFreq'].value_counts().sort_index()
print("\nBin Distribution (should be more balanced):")
print(ef_load_counts)

efficiency_quantiles = pd.qcut(df_engineered['Study_Efficiency'], q=n_bins, retbins=True, duplicates='drop')[1]
load_quantiles = pd.qcut(df_engineered['Study_Load'], q=n_bins, retbins=True, duplicates='drop')[1]

print("\n\nBin Boundaries (Equal-Frequency):")
print(f"\nStudy_Efficiency bins: {np.round(efficiency_quantiles, 3)}")
print(f"Study_Load bins: {np.round(load_quantiles, 3)}")
```

**Benefits:**

2. **Tree-based Model Improvement:** Decision trees work well with binned features
3. **Handles Outliers:** Extreme values are grouped with others in same bin

**Drawbacks:**
1. **Information Loss:** Converts continuous information to categorical
2. **Loss of Precision:** Similar values in same bin are treated identically

4. **Arbitrary Boundaries:** Equal-width method's boundaries may not match data distribution

#### 3.1 D1. Measures of Similarity and Dissimilarity


```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([2, 3, 4, 5])

euclidean = np.linalg.norm(a - b)

cosine = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

pearson = np.corrcoef(a, b)[0, 1]

A = {1, 2, 3}
B = {2, 3, 4}
jaccard_sets = len(A & B) / len(A | B)

universe = sorted(list(A | B))
vecA = np.array([1 if x in A else 0 for x in universe])
vecB = np.array([1 if x in B else 0 for x in universe])
jaccard_binary = np.logical_and(vecA, vecB).sum() / np.logical_or(vecA, vecB).sum()

print('Euclidean distance (a, b):', euclidean)
print('Cosine similarity (a, b):', cosine)
print('Pearson correlation (a, b):', pearson)
print('Jaccard similarity (sets A,B):', jaccard_sets)
print('Jaccard similarity (binary vectors):', jaccard_binary)
```

Metric explanations

- Euclidean distance: Straight-line distance between two numeric vectors in Euclidean space; lower means more similar in absolute terms.

- Cosine similarity: Measures angle between vectors (direction); ranges from -1 to 1, 1 means identical direction regardless of magnitude.

- Pearson correlation: Linear correlation coefficient; measures linear relationship and is scale-invariant (centered by mean).

- Jaccard similarity: For sets or binary attributes, ratio of intersection over union; useful for presence/absence comparisons. For example documents

```python
print(' Comparison notes ')
print('Euclidean vs Cosine: Euclidean depends on absolute differences; cosine focuses on direction (relative proportions).')
print('Example: if you scale a vector by a positive constant, cosine remains the same but Euclidean changes.')
print('')
print('Pearson vs Cosine: Both ignore scale differences to some extent (Pearson subtracts mean, cosine does not), but Pearson centers the data so it measures linear co-movement.')
print('')
print('Jaccard vs numeric metrics: Jaccard is for sets/binary data and ignores magnitude â€” it only cares about shared presence. Numeric metrics require numeric features and capture magnitude/direction.')

a_scaled = a * 10
eu_scaled = np.linalg.norm(a_scaled - b)
cos_scaled = np.dot(a_scaled, b) / (np.linalg.norm(a_scaled) * np.linalg.norm(b))
print('\nScaling a by 10 -> Euclidean changes, Cosine stays similar:')
print('Euclidean (scaled a, b):', eu_scaled)
print('Cosine (scaled a, b):', cos_scaled)

results = {
    'euclidean': float(np.linalg.norm(a - b)),
    'cosine': float(cosine),
    'pearson': float(pearson),
    'jaccard_sets': float(jaccard_sets),
    'jaccard_binary': float(jaccard_binary)
}
```

#### 3.2 D2. Dimensionality Reduction (PCA) 


```python
import numpy as np
import matplotlib.pyplot as plt

num_df = df.select_dtypes(include=[np.number]).copy()
num_df = num_df.loc[:, num_df.std(ddof=0) != 0]
X_orig = num_df.fillna(num_df.mean())

X = X_orig.values.astype(float)
mean = X.mean(axis=0)
std = X.std(axis=0, ddof=0)
X_std = (X - mean) / std

cov = np.cov(X_std, rowvar=False)

eigvals, eigvecs = np.linalg.eigh(cov)
idx = np.argsort(eigvals)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]

explained_ratio = eigvals / eigvals.sum()
print('Eigenvalues (desc):', np.round(eigvals, 6))
print('Explained variance ratio:', np.round(explained_ratio, 6))

W = eigvecs[:, :2]
X_pca = X_std.dot(W)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], s=40, alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA projection (2 components)')
plt.grid(True)
plt.show()

print('Original numeric shape:', X.shape)
print('Projected shape:', X_pca.shape)
```

PCA steps 

- Standardize features: Center each feature to mean=0 and scale to unit variance so features contribute equally.
- Covariance matrix: Measures how features vary together; PCA finds directions of maximum variance.
- Eigenvalues/eigenvectors: Eigenvectors are principal directions; eigenvalues show the variance captured along each direction.
- Select top eigenvectors: Choose components with largest eigenvalues (most variance).
- Projection: Represent original data in the new lower-dimensional basis (principal components).

**What eigenvalues represent:** larger eigenvalues = more variance captured by the corresponding principal component.

**What the PCA plot shows:** a 2D view of the data in the directions of maximum variance â€” clusters or spread visible here reflect dominant structure; some original detail is lost but main variance directions are preserved.
